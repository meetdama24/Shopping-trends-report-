```python
!pip install pandas matplotlib seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Retail Sales Dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

# Show first 5 rows
df.head()
# Shape of dataset
print("Rows & Columns:", df.shape)

# Data types
print("\nData Types:\n", df.dtypes)

# Summary statistics
print("\nSummary Statistics:\n", df.describe())

# Missing values
print("\nMissing Values:\n", df.isnull().sum())
# 1. Distribution of Total Bills
plt.figure(figsize=(8,5))
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title("Total Bill Distribution")
plt.show()

# 2. Average Tip by Day
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="day", y="tip", estimator="mean")
plt.title("Average Tip by Day")
plt.show()

# 3. Total Bill vs Tip
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", size="size")
plt.title("Total Bill vs Tip")
plt.show()

# 4. Tips by Smoking Status
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="smoker", y="tip")
plt.title("Tips by Smoker/Non-Smoker")
plt.show()
avg_tip = df['tip'].mean()
best_day = df.groupby('day')['tip'].mean().idxmax()
best_tip = df.groupby('day')['tip'].mean().max()

print("ðŸ“Š Insights:")
print(f"- Average tip amount is ${avg_tip:.2f}.")
print(f"- Highest average tips are on {best_day} (${best_tip:.2f}).")
print("- Larger bills generally lead to larger tips.")
print("- Non-smokers tend to tip slightly more than smokers.")
```
